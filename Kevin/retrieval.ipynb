#Cleaning Data

```python
df2.columns
```

```python
df2 = df2.replace({'income': {99999999: np.nan}})
df = df2
```

```python
df = df.replace({'marital_status': {'Never married/single':6,'Married spouse present':1,'Divorced':4,'Widowed':5,
                                    'Separated':3,'Married spouse absent':2},'gender' : {'Male':1, 'Female':0}})
```

```python
df = df.dropna(axis=0,subset=['income'])
```

```python
df_employed = df.loc[df['income'] != 0]
```

```python
household_index = df_employed.groupby(['serial_num'])['income'].idxmax()
total_income = df_employed.groupby(['serial_num'])['income'].sum()
```

```python
df_household = df_employed.loc[household_index]
df_household = df_household.join(total_income, on='serial_num', rsuffix="_total")
df_household.head()
```

# Making Dummies

```python
met_id_dummies = pd.get_dummies(df_household.state_id)
df_with_dummies_met = df_household.join(met_id_dummies, rsuffix="edu")
# df_with_dummies_met = df_with_dummies_met.loc[df_with_dummies_met['met_id'] != 9999]
```

```python
edu_id_dummies = pd.get_dummies(df_household.marital_status)
df_with_dummies_2 = df_with_dummies_met.join(edu_id_dummies, rsuffix="marital")
```

```python
# exp_new = df_with_dummies_met.replace({"edu_level_id" : {2 : 1,10 : 1,20 : 1,30 : 1,40 : 1,50 : 1,60 : 1,70 : 1,71 : 1,73:1}})
# df_with_dummies_met = exp_new.replace({"edu_level_id" : {81:2 ,91:3 ,92:3 ,111:4 ,123:5 ,124:6 ,125:6 }})
```

```python
df_with_dummies_2.head()
```

```python
all_features_met = list(df_with_dummies_2.columns)
all_features_met.remove ("own")
all_features_met.remove ("relationship")
all_features_met.remove ("income")
all_features_met.remove ("serial_num")
all_features_met.remove ("person_weight")
all_features_met.remove ("gender")
all_features_met.remove ("met_id")
all_features_met.remove ("marital_status")
all_features_met.remove ("state_id")

X_met = df_with_dummies_2[all_features_met].astype(int)

# X_met = df_with_dummies_met.met_id_dummies

Y_met = df_with_dummies_met['own']

X_train_met, X_test_met, Y_train_met, Y_test_met = train_test_split(X_met,Y_met)
```

```python
selector = SelectKBest(k=6)
selector.fit_transform(X_met,Y_met)
print selector.get_support(indices=True)
```

```python
feature_selector = LogisticRegression(penalty="l1")
feature_selector.fit_transform(X_train_met,Y_train_met)
X_new_train_met = feature_selector.transform(X_train_met, threshold= 1e-6)
X_new_test_met = feature_selector.transform(X_test_met, threshold= 1e-6)
model_log_met = LogisticRegression(penalty="l1").fit(X_new_train_met, Y_train_met)
Y_prediction_log_met = model_log_met.predict(X_new_test_met)
print feature_selector.coef_
print classification_report(Y_test_met,Y_prediction_log_met)
```
